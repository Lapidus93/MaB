```python
import random


#ФУНКЦИЯ ВЫБОРА РАНДОМНОЙ КАРТЫ ИЗ ЛЮБОЙ КОЛОДЫ

def take_card(deck):
    return random.choice(deck).show_card()



### КЛАСС КОЛОД
#TAKE_CARD - взять и показать рандомную карту из колоды

class deck(object):
    
    def __init__(self, name, cards_id):
        self.name = name
        self.cards_id = cards_id

    def take_card(self):
        card_id = int(random.choice(self.cards_id))
        card_itself = list_of_cards[card_id]
        card_itself.show_card()
    
# КЛАСС КАРТ
#show_card - красиво вывести содержимое карты

class card(object):
    
    def __init__(self, cid, name, description,actions):
        self.cid = cid
        self.name = name
        self.description = description
        self.actions = actions
         
  
    def show_card(self):
        print('##################')
        print('###', self.name)
        print('#########')
        print('#',self.description)
        print('#########')
        print('#', 'Действия:')
        for i in self.actions:
            print('#',i)
        print('##################') 
        
        
### СОЗДАНИЕ КОЛОД

file = open('баланс_колод.txt', 'r', encoding="utf-8")
list_of_strings = []

x = file.readlines()
for i in x:
    y = i.split(';')
    list_of_strings.append(y)
    
list_of_decks = []
for i in list_of_strings:
    dec = deck(i[0],i[1:])
    list_of_decks.append(dec)
    
file.close()


file = open('список_карт.txt', 'r', encoding="utf-8")
x = file.readlines()

list_of_cards = []
#cid, name, description,actions
for i in x:
    i = i.split(';')   
    cid = i[0]
    name = i[1]
    description = i[2]
    actions = i[3].split('-')
    car = card(cid,name,description,actions)
    list_of_cards.append(car)
    
    
    
import random

#КЛАСС - ЗАДАНИЯ
#show_card - красиво показать карту задания

class tasks(object):
    
    def __init__(self, id_t, deck, t_type, diff, length, name, description, to_do, reward):
        self.id_t = id_t
        self.deck = deck
        self.t_type = t_type
        self.diff = diff
        self.length = length
        self.name = name
        self.description = description
        self.to_do = to_do
        self.reward = reward

    def show_card(self):
        print('##################')
        print('###', self.name)
        print('#########')
        print('###', self.t_type)
        print('#########')
        print('###','Награда:', self.reward)
        print('#########')
        print('#',self.description)
        print('#########')
        print('###', 'Как выполнять:')
        for i in self.to_do:
            print('#',i)
        print('##################')
        
        
#СОЗДАЕМ списки-колоды заданий
prav_war = []
prav_not_war = []
starost_not_war = []
startost_war = []
tavern_not_war = []
tavern_war = []
defence_war = []
defence_not_war = []
help_war = []
help_not_war = []
warrior_guild = []



# выгружаем задания из файла и дробим на кусочки
file = open('tasks.txt', 'r', encoding= 'utf-8')
raw_tasks = file.readlines()
file.close()
splited_tasks = []

for i in raw_tasks:
    i = i.split(';')
    splited_tasks.append(i)
    
    
# Преобразовываем строки в ОБЪЕКТЫ ЗАДАНИЙ и распределяем по колодам
tasks_obj = []
for i in splited_tasks:
    separated_task_plan = i[7].split('---')
    task = tasks(i[0], i[1], i[2], i[3], i[4], i[5], i[6], separated_task_plan, i[8] )
    tasks_obj.append(task)
    
    
for i in tasks_obj:
    if i.deck == 'prav_war':
        prav_war.append(i)
        
    elif i.deck == 'prav_not_war':
        prav_not_war.append(i)
        
    elif i.deck == 'starost_not_war':
        starost_not_war.append(i)
        
    elif i.deck == 'startost_war':
        startost_war.append(i)
        
    elif i.deck == 'tavern_not_war':
        tavern_not_war.append(i)
        
    elif i.deck == 'tavern_war':
        tavern_war.append(i)
        
    elif i.deck == 'defence_war':
        defence_war.append(i)
        
    elif i.deck == 'defence_not_war':
        defence_not_war.append(i)
        
    elif i.deck == 'help_war':
        help_war.append(i)
        
    elif i.deck == 'help_not_war':
        help_not_war.append(i)
        
    elif i.deck == 'warrior_guild':
        warrior_guild.append(i)
        
list_of_tasks_decks = []
list_of_tasks_decks.append(prav_not_war)
list_of_tasks_decks.append(prav_war)
list_of_tasks_decks.append(starost_not_war)
list_of_tasks_decks.append(startost_war)
list_of_tasks_decks.append(tavern_not_war)
list_of_tasks_decks.append(tavern_war)
list_of_tasks_decks.append(warrior_guild)

list_of_tasks_decks.append(defence_not_war)
list_of_tasks_decks.append(defence_war)
list_of_tasks_decks.append(help_not_war)
list_of_tasks_decks.append(help_war)


class jobs(object):
    
  #КЛАСС РАБОТЫ (Создаем класс, выгружаем из файла текст и создаем объекты)
     
    def __init__(self, cid, name, description,rank,diff,reward):
        self.cid = cid
        self.name = name
        self.description = description
        self.rank = rank
        self.diff = diff
        self.reward = reward
        
    def show_card(self):
        print('__________________')
        print('##################')
        print('###', self.name)
        print('#########')
        print('###','Награда:', self.reward)
        print('#########')
        print('#',self.description)
        print('#########')
        print('###Сложность:', self.rank,self.diff)
        print('##################')
        print('__________________')
        print('')

# выгружаем задания из файла и дробим на кусочки
file = open('jobs.txt', 'r', encoding= 'utf-8')
raw_jobs = file.readlines()
file.close()
splited_jobs = []

for i in raw_jobs:
    i = i.split(';')
    splited_jobs.append(i)
    
    
# Преобразовываем строки в ОБЪЕКТЫ ЗАДАНИЙ и распределяем по колодам
jobs_obj = []
for i in splited_jobs:
    job = jobs(i[0], i[1], i[2], i[3], i[4], i[5])
    jobs_obj.append(job)

builders = []
healers = []
for i in jobs_obj:
    if i.name == 'Гильдия строителей':
        builders.append(i)
    elif i.name == 'Гильдия лекарей':
        healers.append(i)
```
