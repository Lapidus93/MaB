```python
import random

### КЛАСС КОЛОД

#count_cards - посчитать сколько карт в колоде
#take_card - выбрать из колоды карту и показать ее содержимое красиво с помощью метода show_card из класса CARDS

class deck(object):
    
    def __init__(self, name, cards_id):
        self.name = name
        self.cards_id = cards_id
   
    def count_cards(self):
        return len(self.cards_id)
    
    def take_card(self):
        card_id = int(random.choice(self.cards_id))
        card_itself = list_of_cards[card_id]
        card_itself.show_card()
    
# КЛАСС КАРТ

#show_card - красиво вывести содержимое карты


class card(object):
    
    def __init__(self, cid, name, description,actions):
        self.cid = cid
        self.name = name
        self.description = description
        self.actions = actions
         
  
    def show_card(self):
        print('##################')
        print('###', self.name)
        print('#########')
        print('#',self.description)
        print('#########')
        print('#', 'Действия:')
        for i in self.actions:
            print('#',i)
        print('##################')
        

    
### СОЗДАНИЕ КОЛОД

file = open('баланс_колод.txt', 'r', encoding="utf-8")
list_of_strings = []

x = file.readlines()
for i in x:
    y = i.split(';')
    list_of_strings.append(y)
    
list_of_decks = []
for i in list_of_strings:
    dec = deck(i[0],i[1:])
    list_of_decks.append(dec)
    
file.close()


file = open('список_карт.txt', 'r', encoding="utf-8")
x = file.readlines()

list_of_cards = []
#cid, name, description,actions
for i in x:
    i = i.split(';')   
    cid = i[0]
    name = i[1]
    description = i[2]
    actions = i[3].split('-')
    car = card(cid,name,description,actions)
    list_of_cards.append(car)
```
